```python
import requests
def fetch_content_from_url(url):
    # Make a GET request to the URL
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        print(f"Successfully fetched the content from {url}")
        # Print the entire response content
        print(response.text)
    else:
        print(f"Failed to fetch the content from {url}. Status code: {response.status_code}")
# Example usage
url = "https://www.python.org/"
fetch_content_from_url(url)
```

```python
import requests
def download_image(url, file_path):
    # Make a GET request to the URL
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        print(f"Successfully fetched the image from {url}")
        # Write the image content to a file
        with open(file_path, 'wb') as file:
            file.write(response.content)
        print(f"Image saved to {file_path}")
    else:
        print(f"Failed to fetch the image from {url}. Status code: {response.status_code}")
# Example usage
url = "https://www.python.org/static/community_logos/python-logo.png"
file_path = r"python-logo.png"
download_image(url, file_path)
```

```python
# Serialization is the process of converting a Python object into a JSON string.
import json
# Python object (dictionary)
person = {
  "name": "Samrat Paul",
  "age": 25,
  "is_employee": True,
  "courses": ["Python", "Chatbot", "GenAI"],
  "address": {
    "street": "fursungi",
    "city": "Pune",
    "zipcode": "123456"
  }
}
person_json = json.dumps(person)

person_json_indent=json.dumps(obj=person,indent=4)
print(person_json)
print(person_json_indent)
```

```python
def fetch_json_from_url(url):
    # Make a GET request to the URL
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        print(f"Successfully fetched the JSON data from {url}")
        # Parse the JSON content
        data = response.json()
        # Print the JSON data in a pretty format
        print(json.dumps(data, indent=4))
    else:
        print(f"Failed to fetch the JSON data from {url}. Status code: {response.status_code}")
# Example usage
url = "https://jsonplaceholder.typicode.com/users/1"
fetch_json_from_url(url)
```

```python
# Define the endpoint URL
url = "https://jsonplaceholder.typicode.com/posts"
# Define the data to be sent in the POST request
data = {
    "title": "Ankush",
    "body": "To show how we can post data",
    "userId": 100
}
# Make a POST request to the API
response = requests.post(url, json=data)
print(response)
# Check if the request was successful
if response.status_code == 201:
    # Parse the JSON response
    new_post = response.json()
    print("New post created:", new_post)
else:
    print("Failed to create post:", response.status_code)
```
